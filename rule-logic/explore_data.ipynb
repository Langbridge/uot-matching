```python
import pandas as pd
import os
import json
```

```python
with open('../good-instances-v3.txt', 'r') as f:
    instance_list = f.read().splitlines()

instance_list
```

```python
var_dict = {}

with open('../rule-logic-dataset/rules.json', 'r') as rule_file:
    rule_dict = json.load(rule_file)

for inst in instance_list:
    var_dict[inst] = {}

    var_dict[inst]['variables'] = rule_dict[inst.split('_')[0]]['variables']
    with open(f'../rule-logic-dataset/instances/{inst}.json', 'r') as json_file:
        instance_file = json.load(json_file)
    var_dict[inst]['sensors'] = list(instance_file['available_sensors'].keys())
```
